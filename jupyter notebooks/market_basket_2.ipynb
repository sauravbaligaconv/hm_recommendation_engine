```python
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
transaction=pd.read_csv('transactions_train.csv')
articles=pd.read_csv('articles.csv')
```

```python
transaction['t_dat']=pd.to_datetime(transaction['t_dat'], format='%Y-%m-%d')
```

```python
transaction.head()
```

```python
articles.head()
```

```python
articles=articles[['article_id','product_type_name','product_group_name']]
```

```python
articles.head()
```

```python
f_1=pd.merge(left=transaction,right=articles,left_on='article_id',right_on='article_id')
```

```python
f_1.head()
```

```python
f_1.product_type_name.unique()
```

```python
len(f_1)
```

```python
sb=f_1[f_1.product_type_name.isin(['Bikini top', 'Swimwear bottom'])]
```

```python
sb_data=sb.groupby(['customer_id','t_dat'])['article_id'].agg(list).reset_index()
```

```python
sb_data.head()
```

```python
final_data=sb_data.article_id
```

```python
sb_data['unique_prod']=sb_data.article_id.apply(lambda x:set(x))
```

```python
sb_data.head()
```

```python
sb_data['length']=sb_data.unique_prod.apply(lambda x: len(x))
```

```python
sb_data.head()
```

```python
len(sb_data)
```

```python
final_1=sb_data[sb_data.length>1]
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb1=MultiLabelBinarizer()
```

```python
res = pd.DataFrame(mlb1.fit_transform(final_1.unique_prod),
                   columns=mlb1.classes_)
```

```python
res.head()
```

```python
freq_items = apriori(res, min_support = 0.02, use_colnames = True, verbose = 1)
```

```python
freq_items
```

```python
association_rules(freq_items,metric='lift',min_threshold=0.8)
```
