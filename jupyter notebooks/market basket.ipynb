```python
unique_row_items = []
for index, row in data.iterrows():
    items_series = list(row.str.split(','))
    for item_serie in items_series:
        for item in item_serie:
            if item not in unique_row_items:
                unique_row_items.append(item)

unique_row_items
```

```python
unique=[]
for index, row in min_data.iterrows():
    print(index)
    for i in row:
        for j in i:
            if j not in unique:
                unique.append(j)
```

```python
len(unique)
```

```python
df_apriori = pd.DataFrame(columns=unique)
df_apriori
```

```python
b=[]
for index, row in min_data.iterrows():
    one_hot_encoding = np.zeros(len(unique),dtype=int)
    for i in row:
        a=[]
        for j in i:
            a.append(j)
        b.append(a)
```

```python
b
```

```python
df_apriori
```

```python
for it in b:
    for j in it:
        for i,column in enumerate(df_apriori.columns):
            if j == column:
                print(j,column)
                one_hot_encoding[i] = 1
```

```python
for index, row in min_data.iterrows():
    df_apriori.at[index] = one_hot_encoding
```
